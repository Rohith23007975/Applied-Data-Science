```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#load data
path = "/content/drive/MyDrive/Colab Notebooks/dataset/CAR DETAILS FROM CAR DEKHO.csv"
df = pd.read_csv(path)
print(df.head())
row_count = len(df)
print(f'The DataFramenhas {row_count} rows.')

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Handle missing values
df.dropna(inplace=True)

# Descriptive statistics
descriptive_stats = df.describe()
print("Descriptive Statistics:\n", descriptive_stats)

non_numeric_cols = df.select_dtypes(exclude=['number']).columns
print("Non-numeric columns:", non_numeric_cols)

# Handle non-numeric values
# For example, you can encode categorical variables or handle missing values
# For simplicity, we'll just drop non-numeric columns in this example
df_numeric = df.select_dtypes(include=['number'])

# Calculate correlation matrix
correlation_matrix = df_numeric.corr()

# Display correlation matrix
print("Correlation Matrix:")
print(correlation_matrix)

# Create a histogram of a numerical variable
plt.hist(df['year'], bins=15, color='red', edgecolor='black')
plt.title('Distribution of Data')
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.show()

# Create a line graph of trends over time
plt.plot(df['year'], df['km_driven'], color='green')
plt.title('Trends Over Time')
plt.xlabel('year')
plt.ylabel('km_driven')
plt.show()

# Create a correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()



```
